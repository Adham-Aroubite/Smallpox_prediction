```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train_data.csv')
df.head()
```

```python
df["symptome 2"] = df["symptome 2"].astype(int)
df["symptome 3"] = df["symptome 3"].astype(int)
df["symptome 4"] = df["symptome 4"].astype(int)
df["symptome 5"] = df["symptome 5"].astype(int)
df["symptome 6"] = df["symptome 6"].astype(int)
df["symptome 7"] = df["symptome 7"].astype(int)
df["symptome 8"] = df["symptome 8"].astype(int)
df["symptome 9"] = df["symptome 9"].astype(int)
df.head()
```

#### Converting Countries to continents

```python
countries = df['Country']
countries.count()
```

```python
import pycountry_convert as pc

def country_to_continent(country_name):
    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name

# Example
country_name = "Bouvet Island"
print(country_to_continent(country_name))
```

```python
continents = []
for country in countries:
    try:
        result = country_to_continent(country)
        continents.append(result)
    except Exception:
        print(country)
```

```python
continents = pd.DataFrame(continents)
continents = continents.rename(columns={0: 'Continent'})
continents
```

```python
df = pd.concat([df, continents], axis=1)
df
```
